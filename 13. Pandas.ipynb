## Dealing with missing values

- If missing values are in numerical column then we replace them with mean or median of that column

```python
data.price.replace(np.nan, data.price.mean(), inplace = True)
data.horsepower.replace(np.nan, data.horsepower.mean(), inplace = True)
data['peak-rpm'].replace(np.nan, data['peak-rpm'].mean(), inplace  = True)
data.bore.replace(np.nan, data.bore.mean(), inplace = True)
data.stroke.replace(np.nan, data.stroke.mean(), inplace  =True)

data.isnull().sum()
```

- If missing values are in categorical column then we replace them with mode

```python
data['num-of-doors'].replace(np.nan, data['num-of-doors'].mode()[0], inplace = True)

data.isnull().sum()
```

- If number of missing values in a column is large wrt total then the column can be dropped

```python
data.drop('normalized-losses', axis = 1, inplace = True)
```

```python
data.isnull().sum()
```

- If number of all missing values is small then those rows can be removed
                - data.dropna(inplace = True)
