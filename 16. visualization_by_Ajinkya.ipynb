```python
# 5.Encoding
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for col in df.columns:
  if df[col].dtype =='object':
    df[col]=le.fit_transform(df[col])

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Types of plot
# 1.Barplot
# 2.line plot
# 3.scatter plot
# 4.pie charts / pie plot
# 5,histogram
# 6.Boxplot
# 7.heatmap
# 8.area plot
# 9.voilin plot
```

```python
#Line plot
days=['Monday','Tuesday','Wednesday','Thursday','Friday']  #X_axis

values=[450,200,600,500,700]     #y-axxis

plt.plot(days,values,marker='o',color='red')  #for drawing the line
plt.grid(True)
plt.xlabel("Days")
plt.ylabel("Values")
plt.title("Line plot for Days vs values")
plt.show()
```

```python
days=['Monday','Tuesday','Wednesday','Thursday','Friday']  #X_axis

values=[450,200,600,500,700]     #y-axxis

plt.plot(days,values,marker='o',linestyle='dashed',color='red',markerfacecolor='yellow')  #for drawing the line
plt.grid(True)
plt.xlabel("Days")
plt.ylabel("Values")
plt.title("Line plot for Days vs values")
plt.show()
```

```python
#Barplot

speed=[100,200,300,400]
carname=['Alto','Audi','Swift','BMW']

plt.bar(carname,speed,width=0.2,color=['red','black','green','pink'])
plt.grid(True)
plt.show()
```

```python
# piechart
values=[10,30,40,10,10]

category=['children 0-18','Adult 18-25','Adult 25 - 50','Adult 50-75','Senior citizen']


plt.pie(values,labels=category,autopct='%0.2f%%',shadow=True)


plt.show()
```

```python
# piechart
values=[10,30,40,10,10]

category=['children 0-18','Adult 18-25','Adult 25 - 50','Adult 50-75','Senior citizen']


plt.pie(values,labels=category,autopct='%0.2f%%',shadow=True,explode=[0,0.5,0,0.1,0])

plt.legend(title="Category",loc="lower left")
plt.show()
```

```python
# piechart
values=[10,30,40,10,10]

category=['children 0-18','Adult 18-25','Adult 25 - 50','Adult 50-75','Senior citizen']


plt.pie(values,labels=category,autopct='%0.2f%%',shadow=True,colors=['red','grey','pink','green','darkblue'])

plt.show()
```

```python
#Scatter plot-

rollno=[5,7,8,10,12,15]
math=[34,45,55,23,78,66]
science=[12,89,45,75,56,99]
plt.scatter(rollno,math,color='red',marker='*',label="Maths",s=300)
plt.scatter(rollno,science,color="yellow",marker="o",label="science",s=200)
plt.title("Scatter plot between marks nad roll no")
plt.xlabel("roll no of students")
plt.ylabel("Marks")
plt.legend(title="Marks",loc="lower right")
plt.grid(True)
plt.show()
```

```python
#voilin plot

group=["Group-A","Group-A","Group-B","Group-B","Group-C","Group-C","Group-D","Group-D"]

values=[34,45,47,56,59,67,89,69]

sns.violinplot(x=group,y=values)
plt.show()
```

```python
#Histogram
age=[34,45,34,23,56,78,23,25,33,78,12,11,10,18,45]

plt.hist(age,color='red',edgecolor='black',bins=4)
plt.xlabel("Age group")
plt.ylabel("Count of data")
plt.grid(True)
plt.savefig('histplt.png')
plt.show()
```
